```python
def calc_emission_prob(states, observations, mean, variance):
    emission_prob = {}

    for i in range (len(states)):
            emission_prob[i+1] = {}


    for i in range (len(states)):
        for j in range (len(observations)):

            emission_prob[i+1][j+1] = (1/((math. sqrt(variance[i]))* math. sqrt(2 * math.pi)))* math.exp(-((observations[j] - mean[i])**2)/(2*variance[i]))
            
    return emission_prob
```

```python
def step_choose(v):
    hidden_states = []
    max_prob = - 10000000
    best_state = 0
    
    for item in v[-1]:
        if v[-1][item]["probability"] > max_prob:  #log nite hobeeeee, naile loop e dhukbe na
            max_prob = v[-1][item]["probability"]
            best_state = item
#             print("hello")
    hidden_states.append(best_state)
    previous = best_state

    
    for i in range(len(v) - 2, -1, -1):  #backtracking

        hidden_states.insert(0, v[i + 1][previous]["previous"])
        previous = v[i + 1][previous]["previous"]
        
    
    return hidden_states
    
```

```python
def viterbi(obs, sts, sp, tp, ep):
    
    #vterbi forward part
    V = [{}]
    for i in sts:
        V[0][i] = {"probability": np.log(sp[i]) + np.log(ep[i][1]), "previous": 0}
   
    
    for t in range(1, len(obs)):
        V.append({})
            
    for t in range(1, len(obs)):
        
        for i in sts:
            max_tp = V[t - 1][sts[0]]["probability"] + np.log(tp[sts[0]][i])
            prev_st_selected = sts[0]
            for j in sts[1:]:
                tr_prob = V[t - 1][j]["probability"] + np.log(tp[j][i])
                if tr_prob > max_tp:
                    max_tp = tr_prob
                    prev_st_selected = j
                    

            max_prob = max_tp + np.log(ep[i][t+1])
            V[t][i] = {"probability": max_prob, "previous": prev_st_selected}
            
    
    options = step_choose(V)


    print ("The selected hidded states are " )
    print(options)
    
    
    count2 = 0
    count1 = 0
    output = ['0' for i in range(len(options))]
    
    for i in range (len(options)):
        if (options[i] == 2):
            count2 = count2 + 1
            output[i] ="\"La Nina\"\n"
        elif (options[i] == 1):
            count1 = count1 + 1
            output[i] = "\"El Nino\"\n"
    print(count1, count2)
    print(output)
    
    file1 = open('Output.txt', 'w')
    # Writing multiple strings
    # at a time
    file1.writelines(output)

    # Closing file
    file1.close()
    
            
#     print (" with highest probability of %s" % max_prob)

```

# Baum-Welch

```python
def forward(obs, sts, sp, tp, ep):
    
    #forward part
    V = [{}]
    
    s = 0.0
    for i in sts:
        V[0][i] = {"probability": sp[i] * ep[i][1]}
        s = s + V[0][i]["probability"]
    for i in sts:
        V[0][i]["probability"] = V[0][i]["probability"]/s
   
    
    for t in range(1, len(obs)):
        V.append({})
        
    summation = 0.0
    norm_sum = 0.0
    
            
    for t in range(1, len(obs)):
        
       
        for i in sts:
           
            for j in sts:
                tr_prob = V[t - 1][j]["probability"] * tp[j][i] * ep[i][t+1]
#                 print(tr_prob)
                summation  = summation  + tr_prob
#                 print(summation)
                
                    

#             max_prob = tr_prob  
            
            V[t][i] = {"probability": summation}
            norm_sum = norm_sum + V[t][i]["probability"]
            summation = 0.0
        
        for i in sts:
            V[t][i]["probability"] = V[t][i]["probability"]/norm_sum
            
        norm_sum = 0.0
        
            
            
            
            
    return V
            
    
   
```

```python
def backward(obs, sts, sp, tp, ep):
    
    #forward part
    beta = [{}]
    last_index = len(obs)
    
    
   #     s = 0.0
    
    for t in range(1, len(obs)):
        beta.append({})
        
    for i in sts:
        beta[last_index-1][i] = {"probability": 1.0}
#         s = s + beta[last_index-1][i]["probability"]
        
#     for i in sts:
#         beta[last_index-1][i]["probability"] = beta[last_index-1][i]["probability"]/s
        
   
    
    
        
    summation = 0.0
    norm_sum = 0.0
            
    for t in range(len(obs)-2, -1, -1):
        
        
        for i in sts:
           
            for j in sts:
                tr_prob = beta[t + 1][j]["probability"] * tp[i][j]* ep[j][t+1]
#                 print(tr_prob)
                summation  = summation  + tr_prob
#                 print(summation)
                
                    

#             max_prob = tr_prob  
            
            beta[t][i] = {"probability": summation}
            norm_sum = norm_sum + beta[t][i]["probability"]
            summation = 0.0
            
        for i in sts:
            beta[t][i]["probability"] = beta[t][i]["probability"]/norm_sum
            
        norm_sum = 0.0
            
            
    return beta
            
    
   
```

```python
def pi_star_calc(obs,sts,beta,alpha):
    pi_star = [{}]

    for t in range(1, len(obs)):
            pi_star.append({})


    norm_sum = 0.0

    for t in range(0, len(obs)):


        for i in sts:

            pi_star[t][i] = {'probability': alpha[t][i]["probability"] * beta[t][i]["probability"] }



            norm_sum = norm_sum + pi_star[t][i]["probability"]


        for i in sts:
            pi_star[t][i]["probability"] = pi_star[t][i]["probability"]/norm_sum

        norm_sum = 0.0
        
    return pi_star




```
